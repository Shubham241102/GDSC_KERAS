```python
plt.figure(figsize=(6,5))


ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
sns.distplot(Y, hist=False, color="b", label="Fitted Values" , ax=ax1)


plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()
```

# Measures for In-Sample Evaluation

Let's calculate the R^2 for linear regression

```python
#highway_mpg_fit
lr.fit(X, Y)
# Find the R^2
print('The R-square is: ', lr.score(X, Y))
```

```python
Y=lr.predict(X)
print('The output of the first four predicted value is: ', Y[0:4])
```

Now R^2 for Multiple Linear Regression
